Static Attribute:<br>
It applies to the class.<br>
Is called through class name.

```python
class user:
    user_count = 0
    def __init__(self, username, email):
        self.username = username
        self.email = email
        user.user_count += 1

    def display_user(self):
        print(f"username: {self.username}, email: {self.email}")
        print(user.user_count)

user1 = user("david", 'david@yahoo.com')
user2 = user("John", 'john123@gmail.com')

print(user.user_count)
print(user1.user_count)
print(user2.user_count)
user1.display_user()
user2.display_user()
```

@staticmethod decorator:<br>
static + method <br>
for class domain, as utility function

```python
class BankAccount:
    def __init__(self, owner, balance = 0):
        self.owner = owner
        self._balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount      
            print(f"{self.owner}'s new balance ${self._balance}")
        else:
            print("Error: Balance must be positive !")

    @staticmethod #static method for class, called through class name
    def is_valid_interest_rate(rate):
        return 0<= rate <= 10    
    
owner1 = BankAccount("David", 500)
owner1.deposit(200)
print(BankAccount.is_valid_interest_rate(3))
print(BankAccount.is_valid_interest_rate(11))



```
