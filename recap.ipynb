Loops

```python
# for loop .. termination condition is known

for x in range(10):
    print(x, end=" ")

print()

for x in range(1,10):
    print(x,end=" ")

lists = [10,20,30,40]
for list in lists:
    print(list)

# while loop ... termination condition is unknown

x=0
while (x < 10):
    print(f'x is executing. value of x is {x}')
    x = x + 1

# nested loops

for i in range (5):
    for j in range (5):
        print(f"The value of i is {i} and the value of j is {j}")

# Transfer statement
# break
# continue

mylist = [10,20,30,40,50,60]
for list in mylist:
    if list == 50:
        pass
    print(list)


for list in mylist:
    if list == 50:
        pass
    print(list)
else:
    print("this is else block for for loop")

```

Types of arguments in function <br>
<ul>
<li>Positional Argument</li>
<li>Keyword Argument</li>
<li>variable length argument (args*, kwargs**)</li>
<li>Default argument</li>
</ul>

```python
def sum(a,b):   #positional argument
    print("The value of a is ",a)
    print("The value of b is ", b)
    
sum(10,20)

print()

def func1(b = 10, a = 20):   # keyword argument
    print("The value of a is ",a)
    print("The value of b is ", b)

func1()
print()

def func2(a, b=6):   # default argument
    print("The value of a is ",a)
    print("The value of b is ", b)

func2(3)
print()

def func3(*args):   # variale length argument  -- *args
    print(args)
    for arg in args:
        print(arg)

func3(3,4,5,6,7,78,9)


def func4(**kwargs):   # variale length argument -- **kwargs

    print(kwargs.keys())
    print(kwargs.values())

    for k,v in kwargs.items():
        print(f"Value of {k} is {v}")
        


func4(a=3,b=4,c=5,d=6,e=7.9,f=78,g=9)
```

decorators

```python
def decorator(function):
    def decorate(myname):
        print("Good morning")
        print(function(myname))
        print("End of decorator")
    return decorate

@decorator
def name(myname):
     return myname
name('david')
```

Exception Handling

```python
# print("Hello")
# print(10/0)   // error
# print("world")   /// prevents this line 
try:
    print("Hello")
    a = int(input("Enter an integer: "))
    print(10/a)
except (ZeroDivisionError, ValueError):
    print("sorry mate. you cannot divide a number by zero or the input must be integer")

finally:
    print("world")
    print("Code execution was successfull") 
```

```python
class TooYoungException(Exception):
    def __init__(self, msg):
        self.msg = msg

class TooOldException(Exception):
    def __init__(self, msg):
        self.msg = msg

age = int(input("Enter your age: "))

try:
    if age < 18:
        raise TooYoungException("You are too young.")
    elif age > 60:
        raise TooOldException("You are too old.")
    else:
        print("Welcome to dating app.")

except TooOldException:
    print("You are too old for dating app.")

except TooYoungException:
    print("You are too young for dating app.")

finally:
    pass
```
